# Linear Regression to determine price of pizza

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#to determine relation between size of pizza to its price
```

```python
X=np.array([[6],[8],[10],[14],[16],[18],[20]])
```

```python
#X:size of pizza
```

```python
#y:price of pizza
```

```python
y=[70,88,108,151,175,193,200]
```

```python
plt.figure()
```

```python
plt.title('Pizza price plotted against diameter')
```

```python
plt.xlabel('Diameter in inches')
```

```python
plt.ylabel('Price of Pizza in Rupees')
```

```python
plt.plot(X,y,'k.')
```

```python
plt.axis([0,25,0,250])
```

```python
plt.grid(True)
```

```python
#we can see that the relation between price and diameter is linear from the plot
```

```python
plt.show()
```

```python
#model using linear-regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
#predict the price of pizza which is never been predicted.
```

```python
test_pizza=np.array([[24]])
```

```python
predicted_price=model.predict(test_pizza)[0]
```

```python
predicted_price
```

```python
#linear regression class is an estimator all estimator implement fit and predict
```

```python
#determining residual sum of squares
```

```python
print('RSS:%.2f'%np.mean((model.predict(X)-y)**2))
```

```python
#RSS helps to assess the fitness of the function.
```

```python
#Checking the variance of training set
```

```python
var=np.var(X,ddof=1)
```

```python
print(np.var(X,ddof=1))
```

```python
#ddof=1 can be used to set Bessel's correction to calculate sample variance.
```

```python
#finding the covariance of x,y
```

```python
y=np.array([70,88,108,151,175,193,200])
#y is converted to numpy array to ease calculation of mean
```

```python
y_bar=y.mean()
```

```python
x_bar=X.mean()
```

```python
x_bar
```

```python
#covariance formula
```

```python
cov=np.multiply((X-x_bar).transpose(),y-y_bar).sum()/X.shape[0]-1
```

```python
cov
```

```python
#y=alpha+beta*x linear regression model
```

```python
beta=cov/var
```

```python
beta
```

```python
alpha=y_bar-beta*x_bar
```

```python
alpha
```
